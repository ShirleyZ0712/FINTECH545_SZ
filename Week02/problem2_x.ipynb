```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
file = 'problem2_x.csv'
problem2_x_data = pd.read_csv(file)

mean_X = np.mean(problem2_x_data, axis=0)
cov_X = np.cov(problem2_x_data.T)

X1_vals = problem2_x_data['x1'].values
X2_vals = problem2_x_data['x2'].values

conditional_means = mean_X[1] + cov_X[1, 0] / cov_X[0, 0] * (X1_vals - mean_X[0])
conditional_vars = cov_X[1, 1] - (cov_X[1, 0]**2 / cov_X[0, 0])
conditional_std = np.sqrt(conditional_vars)

plt.figure(figsize=(10, 6))
plt.plot(X1_vals, X2_vals, 'o', label='Observed X2')
plt.plot(X1_vals, conditional_means, label='Conditional Mean of X2 | X1', color='red')
plt.fill_between(X1_vals, conditional_means - 1.96 * conditional_std, conditional_means + 1.96 * conditional_std, color='red', alpha=0.3, label='95% Confidence Interval')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.title('Conditional Distribution of X2 given X1')
plt.show()
```
